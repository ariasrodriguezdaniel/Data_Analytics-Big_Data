```python
# Crear una función para definir aquellos que son niños (menores de 16 años)
def male_female_child(Cliente):
    AGE, SEX = Cliente
    
    if AGE < 16:
        return 'child'
    else:
        return SEX
```

```python
df['PERSON'] = df[['AGE', 'SEX']].apply(male_female_child, axis=1)
```

### kde plot, Distribución de las edades de los clientes

#### Agrupar por genero

```python
fig = sns.FacetGrid(df, hue='SEX', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
oldest = df['AGE'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribución de la edad agrupada por género')
fig.add_legend()
```

```python
fig = sns.FacetGrid(df, hue='PERSON', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
oldest = df['AGE'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
fig = sns.FacetGrid(df, hue='SEX', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
oldest = df['AGE'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribución de la edad agrupada por SEXO')
fig.add_legend()
```
